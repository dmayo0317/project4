```python
results = seasonal_decompose(ETHsept_df['Close'],model ='multiplicative',period=12)
plt.rcParams.update({'figure.figsize': (15,10)})
results.plot()
```

```python
#Arima
import warnings
warnings.filterwarnings("ignore")

# Fit auto_arima function to ETH dataset
stepwise_fit = auto_arima(ETHsept_df['Close'], start_p = 0, start_q = 0,
max_p = 4, max_q = 4, m = 12,
start_P = 0, seasonal = True,
d = None, D = 1, trace = True,
error_action ='ignore', # we don't want to know if an order does not work
suppress_warnings = True, # we don't want convergence warnings
stepwise = True)# set to stepwise

# To print the summary
stepwise_fit.summary()
```

```python
# Split data into train / test sets
train = ETHsept_df.iloc[:len(ETHsept_df)-12]
test = ETHsept_df.iloc[len(ETHsept_df)-12:] # set one year(12 months) for testing

# Fit a SARIMAX(0, 2, 1)x(0, 1, 1, 12) on the training set
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(train['Close'],
order = (0, 2, 1),
seasonal_order =(0, 1, 1, 12))

result = model.fit()
result.summary()
```

```python
#Predictions of ARIMA Model agains the Test set
start = len(train)
end = len(train) + len(test) - 1

# Predictions for one-year against the test set
predictions = result.predict(start, end,typ = 'levels').rename("Close")

# plot predictions and actual values
predictions.plot(legend = True)
test['Close'].plot(legend = True)
```

```python
#Evaluate the model using MSE and RMSE
# Load specific evaluation tools
from sklearn.metrics import mean_squared_error
from statsmodels.tools.eval_measures import rmse

# Calculate root mean squared error
rmse(test["Close"], predictions)
```

```python
# Calculate mean squared error
mean_squared_error(test["Close"], predictions)
```

```python
# Train the model on the full dataset
model = model = SARIMAX(ETHsept_df['Close'],order = (0, 2, 2),seasonal_order =(0, 1, 1, 12))
result = model.fit()

# Forecast for the next 3 years
forecast = result.predict(start = len(ETHsept_df),end = (len(ETHsept_df)-1) + 3 * 12,typ = 'levels').rename('Forecast')

# Plot the forecast values
ETHsept_df['Close'].plot(figsize = (20, 10), legend = True)
forecast.plot(legend = True)
plt.title("Ethereum Future Forcast after September", fontsize = 25)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Closing Amount (USD)',fontsize=15)
```
